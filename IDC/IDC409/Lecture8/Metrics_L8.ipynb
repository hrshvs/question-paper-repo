**Triangle inequality (metric):** the dashed side $d(x,z)$ is shorter than or equal to the sum of the other two sides.

```python
import numpy as np
import matplotlib.pyplot as plt

def euclidean_distance(p1, p2):
    #Calculates the Euclidean distance between two points in a 2D plane.

    return np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)



def verify_triangle_inequality(A, B, C):

    # Calculate the three required distances
    dist_AC = euclidean_distance(A, C)
    dist_AB = euclidean_distance(A, B)
    dist_BC = euclidean_distance(B, C)

    # The sum of the two shorter sides
    sum_of_sides = dist_AB + dist_BC


    print("--- Verifying Triangle Inequality ---")
    print(f"Point A: {A}")
    print(f"Point B: {B}")
    print(f"Point C: {C}\n")

    print(f"Distance d(A, C) = {dist_AC:.4f}")
    print(f"Distance d(A, B) = {dist_AB:.4f}")
    print(f"Distance d(B, C) = {dist_BC:.4f}\n")

    print(f"Inequality Check: d(A, C) <= d(A, B) + d(B, C)")
    print(f"Is {dist_AC:.4f} <= {dist_AB:.4f} + {dist_BC:.4f} ")
    print(f"Is {dist_AC:.4f} <= {sum_of_sides:.4f} \n")

    # Check if the inequality holds
    if dist_AC <= sum_of_sides:
        print("SUCCESS: The Triangle Inequality holds true.")
        if np.isclose(dist_AC, sum_of_sides):
            print("   -> Note: The points are collinear, so d(A,C) = d(A,B) + d(B,C).")
    else:
        print("FAILURE: The Triangle Inequality does NOT hold.")


    points = np.array([A, B, C])
    plt.figure(figsize=(8, 6))

    plt.scatter(points[:, 0], points[:, 1], s=100, c=['blue', 'red', 'green'], zorder=5)

    plt.text(A[0] + 0.1, A[1] + 0.1, 'A', fontsize=14)
    plt.text(B[0] + 0.1, B[1] + 0.1, 'B', fontsize=14)
    plt.text(C[0] + 0.1, C[1] + 0.1, 'C', fontsize=14)

    plt.plot([A[0], B[0]], [A[1], B[1]], 'r--', label=f'd(A,B) = {dist_AB:.2f}')
    plt.plot([B[0], C[0]], [B[1], C[1]], 'g--', label=f'd(B,C) = {dist_BC:.2f}')
    plt.plot([A[0], C[0]], [A[1], C[1]], 'b-', linewidth=2, label=f'd(A,C) = {dist_AC:.2f}')
    plt.title("triangle inequality visualization", fontsize=16)
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.legend()
    plt.axis('equal')
    plt.show()


```

Example 1: A standard triangle

```python
p1 = (0, 0)
p2 = (3, 4)
p3 = (5, 1)
verify_triangle_inequality(p1, p2, p3)

print("\n" + "="*40 + "\n")
```

**Symmetry (metric):** arrows both ways on a number line show $d(x,y)=d(y,x)$.


```python
import matplotlib.pyplot as plt

x = 5
y = 12

print(f"Point x = {x}")
print(f"Point y = {y}\n")
dist_xy = abs(x - y)
dist_yx = abs(y - x)
print(f"Distance d(x, y) = |{x} - {y}| = {dist_xy}")
print(f"Distance d(y, x) = |{y} - {x}| = {dist_yx}\n")

if dist_xy == dist_yx:
    print("SUCCESS: The symmetry property d(x, y) = d(y, x) holds true.")
else:
    print(" FAILURE: The symmetry property does NOT hold.")

fig, ax = plt.subplots(figsize=(10, 3))
ax.axhline(0, color='black', linewidth=1)
ax.plot(x, 0, 'bo', markersize=10, label=f'Point x = {x}') # 'bo' is blue circle
ax.plot(y, 0, 'ro', markersize=10, label=f'Point y = {y}') # 'ro' is red circle
ax.annotate('', xy=(y, -0.05), xytext=(x, -0.05),
            arrowprops=dict(arrowstyle='<->', color='purple', lw=2))
ax.text((x + y) / 2, -0.15, f'd(x, y) = {dist_xy}',
        ha='center', va='center', color='purple', fontsize=12)
ax.annotate('', xy=(x, 0.05), xytext=(y, 0.05),
            arrowprops=dict(arrowstyle='<->', color='green', lw=2))
ax.text((x + y) / 2, 0.15, f'd(y, x) = {dist_yx}',
        ha='center', va='center', color='green', fontsize=12)
ax.set_xlim(min(x, y) - 5, max(x, y) + 5)
ax.get_yaxis().set_visible(False)
ax.set_title("Symmetry Property Visualization: d(x, y) = d(y, x)", fontsize=14)
ax.set_xlabel("Number Line", fontsize=12)
ax.grid(True, linestyle='--', alpha=0.6)
ax.legend()

plt.show()
```

**Identity & nonnegativity (metric):** identical points have distance $0$;

 any separation gives a nonnegative length.

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = 7
y1 = 7

dist_identical = abs(x1 - y1)

print("--- 1. Identity Property (d(x, y) = 0 <=> x = y) ---")
print(f"Points are identical: x1 = {x1}, y1 = {y1}")
print(f"Distance d(x1, y1) = |{x1} - {y1}| = {dist_identical}")
if dist_identical == 0:
    print("SUCCESS: Distance is 0 for identical points.\n")
else:
    print("FAILURE: Distance is not 0 for identical points.\n")


# --- Non-negativity ---
# d(x, y) >= 0


x2 = 3
y2 = 10


dist_distinct = abs(x2 - y2)

print("--- 2. Non-negativity Property (d(x, y) >= 0) ---")
print(f"Points are distinct: x2 = {x2}, y2 = {y2}")
print(f"Distance d(x2, y2) = |{x2} - {y2}| = {dist_distinct}")
if dist_distinct > 0:
    print("SUCCESS: Distance is positive for distinct points.")
else:
    print("FAILURE: Distance is not positive for distinct points.")


fig, ax = plt.subplots(figsize=(12, 4))
ax.axhline(0, color='black', linewidth=1)
ax.plot(x1, 0, 'go', markersize=12, label=f'x1 = y1 = {x1}') # Green circle
ax.text(x1, 0.1, f'd(x1, y1) = {dist_identical}\n(Identity)',
        ha='center', va='bottom', color='green', fontsize=12, weight='bold')
ax.plot(x2, 0, 'bo', markersize=10, label=f'x2 = {x2}') # Blue circle
ax.plot(y2, 0, 'ro', markersize=10, label=f'y2 = {y2}') # Red circle
ax.annotate('', xy=(y2, -0.1), xytext=(x2, -0.1),
            arrowprops=dict(arrowstyle='<->', color='purple', lw=2))
ax.text((x2 + y2) / 2, -0.2, f'd(x2, y2) = {dist_distinct}\n(Non-negative)',
        ha='center', va='top', color='purple', fontsize=12, weight='bold')
ax.set_xlim(0, 15)
ax.get_yaxis().set_visible(False)
ax.set_title("Identity & Non-negativity Properties of a Metric", fontsize=16)
ax.set_xlabel("Number Line", fontsize=12)
ax.set_xticks(np.arange(0, 16, 1))
ax.grid(True, linestyle=':', alpha=0.7)
ax.legend(loc='upper right')
plt.show()
```

**Pseudometric:** $d((x_1,y_1),(x_2,y_2))=|x_1-x_2|$  vertical pairs with the same $x$ but different $y$ have distance $0$ (points collapse along $x$).


A pseudometric is a relaxation of the standard metric. It follows all the same rules (non-negativity, symmetry, triangle inequality) with one key exception:

The distance between two distinct points can be zero.

The "Identity" property strictly states that d(x, y) = 0 if and only if x = y. A pseudometric weakens this to x = y implies d(x, y) = 0, but the reverse is not necessarily true.

Analogy: "Distance by City"

Imagine a "distance" function where the distance between two people is 0 if they live in the same city, and 1 if they live in different cities.

d(Person A in Chandigarh, Person B in Chandigarh) = 0. This looks like the identity property.

However, Person A and Person B are different people.

So, d(A, B) = 0 but A  B. This violates the identity property of a true metric, but it perfectly describes a pseudometric.

```python
import matplotlib.pyplot as plt
import numpy as np
import math


x1 = 5.2
y1 = 5.8

#Pseudometric
dist1 = abs(math.floor(x1) - math.floor(y1))

print("--- Case 1: Distinct points, Zero distance ---")
print(f"Points: x1 = {x1}, y1 = {y1}")
print(f"Zone for x1: {math.floor(x1)}, Zone for y1: {math.floor(y1)}")
print(f"Pseudometric distance = {dist1}")
print("Result: d(x1, y1) = 0, even though x1 != y1\n")
x2 = 5.5
y2 = 8.7

dist2 = abs(math.floor(x2) - math.floor(y2))

print("--- Case 2: Distinct points, Non-zero distance ---")
print(f"Points: x2 = {x2}, y2 = {y2}")
print(f"Zone for x2: {math.floor(x2)}, Zone for y2: {math.floor(y2)}")
print(f"Pseudometric distance = {dist2}")
print("Result: d(x2, y2) > 0")

fig, ax = plt.subplots(figsize=(12, 4))

for i in range(0, 12):
    if i % 2 == 0:
        ax.axvspan(i, i + 1, color='gray', alpha=0.2, lw=0)
ax.axhline(0, color='black', linewidth=1)
ax.plot(x1, 0, 'go', markersize=10) # Green circle
ax.plot(y1, 0, 'go', markersize=10)
ax.text((x1 + y1) / 2, 0.1, f'd({x1}, {y1}) = {dist1}',
        ha='center', va='bottom', color='green', fontsize=12, weight='bold')
ax.plot(x2, 0, 'bo', markersize=10) # Blue circle
ax.plot(y2, 0, 'ro', markersize=10) # Red circle
ax.annotate('', xy=(y2, -0.1), xytext=(x2, -0.1),
            arrowprops=dict(arrowstyle='<->', color='purple', lw=2))
ax.text((x2 + y2) / 2, -0.2, f'd({x2}, {y2}) = {dist2}',
        ha='center', va='top', color='purple', fontsize=12, weight='bold')
ax.set_xlim(4, 11)
ax.get_yaxis().set_visible(False)
ax.set_title("Pseudometric Visualization", fontsize=16)
ax.set_xlabel("Number Line with 'Zero-Distance' Zones", fontsize=12)
ax.set_xticks(np.arange(4, 12, 1))
ax.grid(True, axis='x', linestyle=':', alpha=0.7)
plt.show()
```

**Quasimetric:**
A quasimetric is a function that measures "distance" but drops the rule of symmetry. This means the distance from point A to point B might be different from the distance from point B to point A. It still must be non-negative, zero only if the points are identical, and obey the triangle inequality.

Think of the "distance" as the amount of effort it takes to walk between two points on a mountain trail.

* Going Uphill (A to B): This is difficult and requires a lot of effort. Let's say the effort is 10 units.

* Going Downhill (B to A): This is much easier. The effort might only be 3 units.


e.g.
 $d(x,y)=\max(0,y-x)$

 direction matters; $d(2,5)=3$ but $d(5,2)=0$.


 > Delivery agent can charge different charge on deliverying heavy products based on the floor (in case of lift issue).

```python
import matplotlib.pyplot as plt
import numpy as np

pos_A = 4  # Start position
pos_B = 10 # End position
height_A = 2
height_B = 8
A = (pos_A, height_A)
B = (pos_B, height_B)

uphill_factor = 1.5
base_dist = np.sqrt((B[0] - A[0])**2 + (B[1] - A[1])**2)

# Distance from A to B (uphill)
# We apply the uphill factor because we are gaining altitude.
dist_AB = base_dist * uphill_factor
print(f"--- Uphill Journey (A to B) ---")
print(f"Quasimetric distance d(A, B) = {dist_AB:.2f}\n")
# Distance from B to A (downhill)
# We use the baseline effort for the easier downhill journey.
dist_BA = base_dist
print(f"--- Downhill Journey (B to A) ---")
print(f"Quasimetric distance d(B, A) = {dist_BA:.2f}\n")

print("--- Verification ---")
if dist_AB != dist_BA:
    print(f" SUCCESS: The quasimetric is not symmetrical ({dist_AB:.2f} != {dist_BA:.2f}).")
else:
    print(" FAILURE: The quasimetric is symmetrical.")
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot([A[0], B[0]], [A[1], B[1]], color='brown', linestyle='--', label='Trail Path')
ax.plot(A[0], A[1], 'bo', markersize=12, label=f'Point A (Low Point)')
ax.plot(B[0], B[1], 'ro', markersize=12, label=f'Point B (High Point)')
ax.annotate(f'Uphill: d(A, B) = {dist_AB:.2f}',
            xy=B, xytext=(A[0] + 2, A[1] + 4.5),
            arrowprops=dict(arrowstyle='-|>', color='red', lw=2.5, shrinkA=5, shrinkB=15),
            ha='center', va='center', color='red', fontsize=12, weight='bold',
            bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="red", lw=1, alpha=0.9))
ax.annotate(f'Downhill: d(B, A) = {dist_BA:.2f}',
            xy=A, xytext=(B[0] - 2, B[1] - 4.5),
            arrowprops=dict(arrowstyle='-|>', color='green', lw=2.5, shrinkA=5, shrinkB=15),
            ha='center', va='center', color='green', fontsize=12, weight='bold',
            bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="green", lw=1, alpha=0.9))
ax.set_xlim(2, 12)
ax.set_ylim(0, 10)
ax.set_title("Quasimetric Visualization (Uphill vs. Downhill Effort)", fontsize=16)
ax.set_xlabel("Horizontal Position", fontsize=12)
ax.set_ylabel("Height / Altitude", fontsize=12)
ax.grid(True, linestyle=':', alpha=0.7)
ax.legend()
ax.set_aspect('equal', adjustable='box')

plt.show()
```

**Semimetric:** $d(x,y)=|x-y|^2$
 symmetric and nonnegative, but triangle inequality can fail.

A semimetric is a distance measure that follows all the rules of a true metric (it's non-negative, symmetric, and zero only for identical points) except for one: it is not required to satisfy the Triangle Inequality.

This means that taking a detour through a third point can sometimes be a "shortcut" compared to the direct path.

Analogy: The Auto from Tribune chowk (A) to IISER Mohali (B)

Imagine the "distance" is the cost to travel between locations.

* Traveling directly between  (A and B) costs Rs 100

* However, the Autos come to Phase 10 and then Phase 10 to IISER Mohali and Auto take Rs 20 if shared.

The direct path d(A, B) costs Rs 100. But the detour d(A, C) + d(C, B) costs only  20 + 20 = 40. Since 100 is not less than or equal to 40, the Triangle Inequality is broken.

```python
import matplotlib.pyplot as plt
import numpy as np

A = (3, 8)
B = (11, 8)
C = (7, 3)
dist_AB = 100
dist_AC = 20
dist_CB = 20

print("--- Semimetric Verification (Violating Triangle Inequality) ---")
print(f"Direct distance d(A, B) = {dist_AB}")
print(f"Detour distance d(A, C) + d(C, B) = {dist_AC} + {dist_CB} = {dist_AC + dist_CB}\n")

print(f"Is {dist_AB} <= {dist_AC + dist_CB}?")
if dist_AB <= (dist_AC + dist_CB):
    print("FAILURE: The Triangle Inequality holds. This is not a good semimetric example.")
else:
    print("SUCCESS: The Triangle Inequality is VIOLATED. This is a semimetric.")

fig, ax = plt.subplots(figsize=(12, 7))
ax.plot(A[0], A[1], 'bo', markersize=12, label='Tribune Chowk')
ax.plot(B[0], B[1], 'ro', markersize=12, label='IISER Mohali')
ax.plot(C[0], C[1], 'go', markersize=15, label='Phase 10', marker='*')
ax.plot([A[0], B[0]], [A[1], B[1]], 'k--', alpha=0.7)
ax.text((A[0] + B[0]) / 2, (A[1] + B[1]) / 2 + 0.5, f'Direct d(A, B) = {dist_AB}',
        ha='center', color='black', fontsize=12, weight='bold',
        bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", lw=1))
ax.plot([A[0], C[0]], [A[1], C[1]], 'm-', alpha=0.8)
ax.text((A[0] + C[0]) / 2, (A[1] + C[1]) / 2, f'd(A, C) = {dist_AC}',
        ha='center', va='center', color='purple', fontsize=12)

ax.plot([C[0], B[0]], [C[1], B[1]], 'm-', alpha=0.8)
ax.text((C[0] + B[0]) / 2, (C[1] + B[1]) / 2, f'd(C, B) = {dist_CB}',
        ha='center', va='center', color='purple', fontsize=12)
ax.text(7, 1, f"Violation: {dist_AB} is NOT <= {dist_AC + dist_CB}",
        ha='center', fontsize=14, color='red', weight='bold',
        bbox=dict(boxstyle="round,pad=0.5", fc="yellow", alpha=0.8))
ax.set_xlim(0, 14)
ax.set_ylim(0, 12)
ax.set_title("Semimetric Visualization (Triangle Inequality Violation)", fontsize=16)
ax.set_xlabel("Position X", fontsize=12)
ax.set_ylabel("Position Y", fontsize=12)
ax.grid(True, linestyle=':', alpha=0.7)
ax.legend()
ax.set_aspect('equal', adjustable='box')

plt.show()
```

**Ultrametric (other):**


An ultrametric is a special kind of metric that satisfies a stronger version of the Triangle Inequality, often called the Strong Triangle Inequality or the Ultrametric Inequality.

The rule is: d(x, z)  max(d(x, y), d(y, z))

This means the distance to a third point is no greater than the maximum of the other two distances. A strange and powerful consequence of this is that in an ultrametric space, all triangles are either isosceles (with two equal long sides) or equilateral.

Analogy: Hierarchical Clusters (e.g., Family Tree)

Imagine the "distance" between two people is determined by their most recent common ancestor.


* Siblings (A, B): They are very close. Their distance is small (e.g., 1).

* Cousin (C): To find a common ancestor between a person (A) and their cousin (C), you have to go further up the family tree. Their distance is larger (e.g., 5).

* The distance between the other sibling (B) and the cousin (C) is the exact same larger distance (5), because they share the same most recent common ancestor.

This creates an isosceles triangle where d(A, C) = d(B, C). The Strong Triangle Inequality holds: d(A, C)  max(d(A, B), d(B, C)) becomes 5  max(1, 5), which is true.

> concept use in different species

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Circle
A = (4, 5)  # Point in Cluster 1
B = (6, 6)  # Another point in Cluster 1
C = (12, 7) # Point in Cluster 2

# Define cluster membership
cluster_A = 1
cluster_B = 1
cluster_C = 2

dist_AB = 1.0 if cluster_A == cluster_B else 5.0
dist_AC = 1.0 if cluster_A == cluster_C else 5.0
dist_BC = 1.0 if cluster_B == cluster_C else 5.0
print("--- Ultrametric Verification (Strong Triangle Inequality) ---")
print(f"Distance within Cluster 1, d(A, B) = {dist_AB}")
print(f"Distance between clusters, d(A, C) = {dist_AC}")
print(f"Distance between clusters, d(B, C) = {dist_BC}\n")
longest_side = dist_AC
sum_of_others_max = max(dist_AB, dist_BC)
print(f"Checking: d(A, C) <= max(d(A, B), d(B, C))")
print(f"Is {longest_side} <= max({dist_AB}, {dist_BC})?")
print(f"Is {longest_side} <= {sum_of_others_max}?")

if longest_side <= sum_of_others_max:
    print("\n SUCCESS: The Strong Triangle Inequality is SATISFIED.")
else:
    print("\n FAILURE: The Strong Triangle Inequality is VIOLATED.")
fig, ax = plt.subplots(figsize=(14, 8))
cluster1_patch = Circle((5, 5.5), 2.5, color='blue', alpha=0.1, label='Cluster 1')
cluster2_patch = Circle((12, 7), 1.5, color='red', alpha=0.1, label='Cluster 2')
ax.add_patch(cluster1_patch)
ax.add_patch(cluster2_patch)
ax.plot(A[0], A[1], 'bo', markersize=12, label='Point A')
ax.plot(B[0], B[1], 'bo', markersize=12, label='Point B')
ax.plot(C[0], C[1], 'ro', markersize=12, label='Point C')
ax.plot([A[0], B[0]], [A[1], B[1]], 'g-', lw=2)
ax.plot([A[0], C[0]], [A[1], C[1]], 'm--', lw=2)
ax.plot([B[0], C[0]], [B[1], C[1]], 'm--', lw=2)
ax.text((A[0] + B[0]) / 2, (A[1] + B[1]) / 2 + 0.3, f'd(A, B) = {dist_AB}', ha='center', color='green', weight='bold')
ax.text((A[0] + C[0]) / 2, (A[1] + C[1]) / 2 + 0.3, f'd(A, C) = {dist_AC}', ha='center', color='purple', weight='bold')
ax.text((B[0] + C[0]) / 2 - 1, (B[1] + C[1]) / 2 + 0.3, f'd(B, C) = {dist_BC}', ha='center', color='purple', weight='bold')
ax.text(9, 2, f"Result: Triangle is Isosceles\n{longest_side} <= max({dist_AB}, {dist_BC})",
        ha='center', fontsize=14, color='darkgreen', weight='bold',
        bbox=dict(boxstyle="round,pad=0.5", fc="lightgreen", alpha=0.8))
ax.set_xlim(0, 16)
ax.set_ylim(0, 10)
ax.set_title("Ultrametric Visualization (Strong Triangle Inequality)", fontsize=16)
ax.set_xlabel("Position X", fontsize=12)
ax.set_ylabel("Position Y", fontsize=12)
ax.grid(True, linestyle=':', alpha=0.7)
ax.legend()
ax.set_aspect('equal', adjustable='box')

plt.show()
```

# Norm
A **norm** is a function that measures the size (length/magnitude) of a vector.

Formally, a norm on a vector space $V$ is a map $\|\cdot\|: V\to [0,\infty)$ such that for all $x,y\in V$ and scalars $\alpha$:

1. **Nonnegativity:** $\|x\|\ge 0$ and $\|x\|=0 \iff x=0$
2. **Homogeneity:** $\|\alpha x\|=|\alpha|\,\|x\|$
3. **Triangle inequality:** $\|x+y\|\le \|x\|+\|y\|$

A norm induces a **metric**: $d(x,y)=\|x-y\|$.



$ L_1$ Norm , Manhattan norm

```python
import matplotlib.pyplot as plt
import numpy as np
A = (3, 3)
B = (9, 7)
print(f"Point A: {A}")
print(f"Point B: {B}\n")
delta_x = abs(B[0] - A[0])
print(f"Horizontal distance (x) = |{B[0]} - {A[0]}| = {delta_x}")
delta_y = abs(B[1] - A[1])
print(f"Vertical distance (y) = |{B[1]} - {A[1]}| = {delta_y}\n")
l1_distance = delta_x + delta_y
print(f"L1 Distance = x + y = {delta_x} + {delta_y} = {l1_distance}")
fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(A[0], A[1], 'bo', markersize=12, label=f'Point A {A}')
ax.plot(B[0], B[1], 'ro', markersize=12, label=f'Point B {B}')
ax.plot([A[0], B[0]], [A[1], A[1]], 'g--', lw=2, label='Horizontal Path (x)')
ax.plot([B[0], B[0]], [A[1], B[1]], 'm--', lw=2, label='Vertical Path (y)')
ax.text((A[0] + B[0]) / 2, A[1] - 0.5, f'x = {delta_x}', ha='center', color='green', fontsize=12)
ax.text(B[0] + 0.5, (A[1] + B[1]) / 2, f'y = {delta_y}', va='center', color='purple', fontsize=12)
ax.set_xlim(0, 12)
ax.set_ylim(0, 10)
ax.set_title("L1 Norm (Manhattan Distance) Visualization", fontsize=16)
ax.set_xlabel("X-axis", fontsize=12)
ax.set_ylabel("Y-axis", fontsize=12)
ax.grid(True, linestyle=':', alpha=0.7)
ax.legend()
ax.set_aspect('equal', adjustable='box')
ax.text(1, 9, f"Total L1 Distance = {l1_distance}",
        ha='left', fontsize=14, color='darkblue', weight='bold',
        bbox=dict(boxstyle="round,pad=0.5", fc="lightblue", alpha=0.8))
plt.show()
```

$L_2$ Norm: Euclidean Norm

```python
import matplotlib.pyplot as plt
import numpy as np
A = (3, 3)
B = (9, 7)
print(f"--- Calculating L2 Norm ---")
print(f"Point A: {A}")
print(f"Point B: {B}\n")
delta_x = B[0] - A[0]
delta_y = B[1] - A[1]
print(f"Horizontal difference (x) = {B[0]} - {A[0]} = {delta_x}")
print(f"Vertical difference (y) = {B[1]} - {A[1]} = {delta_y}\n")
delta_x_squared = delta_x**2
delta_y_squared = delta_y**2
print(f"x squared = {delta_x_squared}")
print(f"y squared = {delta_y_squared}\n")
sum_of_squares = delta_x_squared + delta_y_squared
print(f"Sum of squares = {delta_x_squared} + {delta_y_squared} = {sum_of_squares}\n")
l2_distance = np.sqrt(sum_of_squares)
print(f"L2 Distance = sqrt({sum_of_squares}) = {l2_distance:.4f}")
fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(A[0], A[1], 'bo', markersize=12, label=f'Point A {A}')
ax.plot(B[0], B[1], 'ro', markersize=12, label=f'Point B {B}')
ax.plot([A[0], B[0]], [A[1], A[1]], 'g--', lw=2, label=f'Horizontal component (x = {delta_x})')
ax.plot([B[0], B[0]], [A[1], B[1]], 'm--', lw=2, label=f'Vertical component (y = {delta_y})')
ax.plot([A[0], B[0]], [A[1], B[1]], 'k-', lw=3, label=f'L2 Distance ')
ax.text((A[0] + B[0]) / 2, A[1] - 0.5, f'x = {delta_x}', ha='center', color='green', fontsize=12)
ax.text(B[0] + 0.5, (A[1] + B[1]) / 2, f'y = {delta_y}', va='center', color='purple', fontsize=12)
ax.set_xlim(0, 12)
ax.set_ylim(0, 10)
ax.set_title("L2 Norm (Euclidean Distance) Visualization", fontsize=16)
ax.set_xlabel("X-axis", fontsize=12)
ax.set_ylabel("Y-axis", fontsize=12)
ax.grid(True, linestyle=':', alpha=0.7)
ax.legend()
ax.set_aspect('equal', adjustable='box')
ax.text(1, 9, f"Total L2 Distance = {l2_distance:.2f}",
        ha='left', fontsize=14, color='black', weight='bold',
        bbox=dict(boxstyle="round,pad=0.5", fc="lightgray", alpha=0.8))
plt.show()
```

* Choose **$\ell_1$** for **sparsity** or robustness to small outliers.
* Choose **$\ell_2$** for smoothness, isotropy, and the standard geometry.
* Choose **$\ell_3$** when you want **convex, smooth**, non-sparse behavior that **penalizes big coordinates/errors more** than $\ell_2$ (but less worst-case than $\ell_\infty$).
* Choose **$\ell_\infty$** to cap the **largest** coordinate deviation.


$ f(x) = \exp(x)$  (convex)  

 the chord lies ABOVE the curve at the midpoint (inequality holds)

 Convexity inequality: $f(x + (1 - )y)   f(x) + (1 -)f(y)$


```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-2.5, 2.5, 400)
f = np.exp(x)
x0, x1, lam = -1.2, 1.4, 0.35
z = lam*x0 + (1 - lam)*x1
f0, f1, fz = np.exp(x0), np.exp(x1), np.exp(z)
rhs = lam*f0 + (1 - lam)*f1
plt.figure()
plt.plot(x, f, label='f(x) = exp(x)')
plt.plot([x0, x1], [f0, f1], linestyle='--', label='Chord between (x0,f0) and (x1,f1)')
plt.plot([x0, x1, z], [f0, f1, fz], 'o')
plt.title('Convex function: f(x)=exp(x)\nConvexity: f(z)  f(x0)+(1)f(x1)')
plt.xlabel(f"Check: f(z)={fz:.3f}  vs  RHS={rhs:.3f}    {'Holds' if fz <= rhs + 1e-12 else 'Fails'}")
plt.grid(True)
plt.legend()
plt.show()
```


Nonconvex function: $g(x) = x^4 - 3x^2$  (double well)

For symmetric points around 0, the inequality can FAIL
Convexity can fail: $g(z)  g(x0)+(1 - )g(x1)$ is False

```python
x = np.linspace(-2.5, 2.5, 400)
g = x**4 - 3*x**2
x0, x1, lam = -1.5, 1.5, 0.5
z = lam*x0 + (1 - lam)*x1  # = 0
g0, g1, gz = x0**4 - 3*x0**2, x1**4 - 3*x1**2, 0.0
rhs = lam*g0 + (1 - lam)*g1  # negative here
plt.figure()
plt.plot(x, g, label='g(x) = x^4 - 3x^2')
plt.plot([x0, x1], [g0, g1], linestyle='--', label='Chord between (x0,g0) and (x1,g1)')
plt.plot([x0, x1, z], [g0, g1, gz], 'o')
plt.title('Nonconvex function: g(x)=x^43x^2')
plt.xlabel(f"Check: g(z)={gz:.3f}  vs  RHS={rhs:.3f}    {'Holds' if gz <= rhs + 1e-12 else 'Fails'}")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
x1_nonconv = 0
x2_nonconv = math.pi
y1_nonconv = math.sin(x1_nonconv)
y2_nonconv = math.sin(x2_nonconv)
mid_x_nonconv = (x1_nonconv + x2_nonconv) / 2
mid_y_on_line_nonconv = (y1_nonconv + y2_nonconv) / 2
y_on_curve_nonconv = math.sin(mid_x_nonconv)
plt.figure(figsize=(10, 7))
ax2 = plt.gca()
x_vals_sin = np.linspace(0, math.pi, 100)
y_vals_sin = np.sin(x_vals_sin)
ax2.plot(x_vals_sin, y_vals_sin, 'b-', label='f(x) = sin(x) (Curve)')
ax2.plot([x1_nonconv, x2_nonconv], [y1_nonconv, y2_nonconv], 'k--', label='Line')
ax2.plot(mid_x_nonconv, y_on_curve_nonconv, 'go', markersize=10, label=f'Point on Curve ({mid_x_nonconv:.2f}, {y_on_curve_nonconv:.2f})')
ax2.set_title("Non-Convex Function: Curve is above the Line", fontsize=16)
ax2.set_xlabel("x (radians)")
ax2.set_ylabel("y")
ax2.grid(True, linestyle=':')
ax2.legend()
plt.show()
```
