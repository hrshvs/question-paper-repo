What is difference between flatten and ravel

Main difference is that flatten always returns a new copy of the array, while ravel returns a view of the original array whenever possible.

Changes made to the array returned by ravel might affect the original array, while changes to the one from flatten will not.

````python
import numpy as np

# Create our original 2D array
a = np.array([[1, 2, 3],
              [4, 5, 6]])

print(f"Original array:\n{a}\n")

# --- Using flatten (makes a copy) ---
f = a.flatten()
f[0] = 99  # Change the first element of the flattened array
print("After modifying the 'flattened' array...")
print(f"Flattened array: {f}")
print(f"Original array is UNCHANGED:\n{a}\n")


# --- Using ravel (makes a view) ---
r = a.ravel()
r[0] = 77  # Change the first element of the raveled array
print("After modifying the 'raveled' array...")
print(f"Raveled array:   {r}")
print(f"Original array IS CHANGED:\n{a}")
````

When we create array slices, views from the array are returned rather than copies of the array data.

````python
A = np.random.randint(low=0,high=100,size=(6,6))
print('Original A \n',A)

Sub_A = A[2:]

print('\n sub_A \n', Sub_A)

Sub_A[3,3]=999
Sub_A[3,4]=999
Sub_A[3,5]=999
print('\n Modified sub_A \n', Sub_A)
print('Now the Original A also get modified \n',A)

````

# Normalize array

Normalizing an array is the process of bringing the array values to some defined range.

We can normalize an array so that its value vary from 0 to 1.

$x = \frac{x- x_{min}}{x_{max} - x_{min}}$

    

````python
arr=np.array([-12311,123, 1231, 441,421,4242,123,41312,4242,122,341,341,1234,122343])
print(' Before normalization \n :', arr)
arr_max = arr.max()
arr_min = arr.min()
arrN= (arr- arr_min)/(arr_max-arr_min)
print(' Before normalization \n :', arrN)
````

#Array Subsets

One can easily create one or more subsets of numpy array. These subsets will also be numpy array objects.

One should be careful to have copy.

One can do following
* Create a subsetof array that has only diagonal elements from original array.
* Create a subset having a dimension omitted if there is only one element in tha dimensions
* Subset created on boolean conditions.
* Based on the array of indices.

Some of the methods are **choose(), diagonal(), take(), ...**



````python
arr = np.array([[1,2,3,4],[11,12,13,14],[21,22,23,24],[31,32,33,34]])
print(arr)

##  row 2, row 3, row 1, row 0
print('\n np.choose([2,3,1,0], arr) ')
newarr= np.choose([2,3,1,0], arr)
print(newarr)


print('\n newarr2= np.choose([2,4,1,0], arr, mode=\'clip\'') # 4 goes to 3 (4-1)')
newarr2= np.choose([2,4,1,0], arr, mode='clip') # 4 goes to 3 (4-1)
print(newarr2)

print('\n newarr2= np.choose([2,5,5,0], arr, mode=\'clip\'') # 5 get clipped  to 3')
newarr22= np.choose([2,5,5,0], arr, mode='clip') # 5 gets clipped to 3
print(newarr22)

#'wrap' mode will use your indices modulo the length of the array
print('\n newarr3= np.choose([2,4,1,0], arr, mode=\'wrap\'') # 4 goes to (4 mod 4)')
newarr3= np.choose([2,4,1,0], arr, mode='wrap') # 4 goes to (4 mod 4)
print(newarr3)
print('\n newarr3= np.choose([2,5,1,0], arr, mode=\'wrap\'') # 5 goes to (5 mod 4)')
newarr3= np.choose([2,5,1,0], arr, mode='wrap') # 5 goes to (5 mod 4)
print(newarr3)
````

````python
arr = np.array([[1,2,3,4],[11,12,13,14],[21,22,23,24],[31,32,33,34]])
print(arr)
print('Diagonal')
dia = arr.diagonal()
print(dia)

print(' \n  ----- \n')
arr3D = np.arange(27).reshape(3,3,3)
print(arr3D)
print('Diagonal')
print(arr3D.diagonal(axis1=1, axis2=2))
````
