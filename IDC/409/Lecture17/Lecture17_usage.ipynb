```python
from sympy import *

x, y = symbols('x y')
expr = x**2 + 2 * y + y**3
print("Expression : {}".format(expr))

# Use sympy.Derivative() method
expr_diff = Derivative(expr, x)

print("Derivative of expression with respect to x : {}".format(expr_diff))
print("Value of the derivative : {}".format(expr_diff.doit()))
```

```python

import sympy as sp #symbols, Matrix
x1,x2,x3,x4= sp.symbols('x1 x2 x3 x4')
f= (x1+ 10*x2)**2 + 5*(x3 - x4)**2 + (x2-2*x3)**4 + 10*(x1-x4)**4
hessian = sp.hessian(f,(x1,x2,x3,x4))

print(hessian)


```

```python
F=Matrix([f])
Jacobian_mat = F.jacobian([x1,x2,x3,x4])
print(Jacobian_mat)
```

```python
new_x1 = 0.1393
new_x2 = -0.0139
new_x3 = 0.0222
new_x4 = 0.0222
hessian_matrix_new = hessian.subs({x1: new_x1, x2: new_x2, x3: new_x3, x4: new_x4})
F_new = F.subs({x1: new_x1, x2: new_x2, x3: new_x3, x4: new_x4})
xnew = Matrix([[new_x1], [new_x2], [new_x3], [new_x4]])
print(hessian_matrix_new)
Jacobian_new = Jacobian_mat.subs({x1: new_x1, x2: new_x2, x3: new_x3, x4: new_x4})
print(Jacobian_new)
print(xnew)
print(F_new)
```

```python

H_inv = hessian_matrix_new.inv()
print(H_inv)
Hgproduct = H_inv * Jacobian_new.transpose()
print(Hgproduct)
#print(xnew)
newnewX = xnew - Hgproduct

print(newnewX)
```

```python
from scipy.optimize import minimize

def powell_function(x):
    return (x[0] + 10 * x[1])**2 + 5 * (x[2] - x[3])**2 + (x[1] - 2 * x[2])**4 + 10 * (x[0] - x[3])**4

# Initial guess
x0 = [3, -1, 0, 1]

# Perform the minimization
result = minimize(powell_function, x0, method='Powell')

print("Optimal soln ")
print(result.x)
print("Minimum value of the Powell function ")
print(result.fun)
```

```python
from scipy.optimize import minimize


def our_function(x):
    return (10*x[0]*x[0] + 5 *x[0] *x[1] + 10*(x[1] - 3)**2 )

# Initial guess
x0 = [10,15]

# Perform the minimization
result = minimize(our_function, x0, method='CG')

# Print the result
print("Optimal soln ")
print(result.x)
print("Minimum value of the our function ")
print(result.fun)
```
