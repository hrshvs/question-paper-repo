```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection, Line3DCollection


v1 = np.array([2.0, 1.0, 1.0])
v2 = np.array([1.0, 2.0, 0.5])
v3 = np.array([0.5, 0.5, 2.0])


n = np.cross(v1, v2)                                   # base normal
n_norm = np.linalg.norm(n)
if n_norm == 0:
    raise ValueError("v1 and v2 are collinear; normal/volume undefined.")
n_hat = n / n_norm

signed_volume = float(np.dot(n, v3))                   # (v1 x v2)  v3
volume = abs(signed_volume)
orientation = "RIGHT-handed (+)" if signed_volume >= 0 else "LEFT-handed ()"

#  Parallelogram face (base)
O = np.zeros(3)
P1 = v1
P2 = v2
P12 = v1 + v2
base_face = [O, P1, P12, P2]
edges = [
    [O, P1], [O, P2], [P1, P12], [P2, P12],
    [O, P12]
]

# Normal arrow drawn from face center
face_center = (O + P1 + P12 + P2) / 4.0
# Choose a pleasant normal length relative to v1,v2 sizes
scale = 0.5 * min(np.linalg.norm(v1), np.linalg.norm(v2))
normal_vec = n_hat * scale

# Orientation arc in the v1-v2 plane (from v1 toward v2)
# Build an orthonormal basis (e1 along v1, e2 in the plane toward v2)
e1 = v1 / np.linalg.norm(v1)
v2_perp = v2 - np.dot(v2, e1) * e1
if np.linalg.norm(v2_perp) < 1e-12:
    # v2 parallel to v1  -> arc is degenerate; draw a short line instead
    arc_pts = np.vstack([np.zeros(3), 0.2 * e1])
else:
    e2 = v2_perp / np.linalg.norm(v2_perp)
    theta = np.arccos(
        np.clip(np.dot(v1, v2) / (np.linalg.norm(v1)*np.linalg.norm(v2)), -1.0, 1.0)
    )
    r = 0.4 * min(np.linalg.norm(v1), np.linalg.norm(v2))
    ts = np.linspace(0, theta, 60)
    arc_pts = np.array([r * (np.cos(t)*e1 + np.sin(t)*e2) for t in ts])


fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

# Vectors
ax.quiver(0,0,0, *v1, color='red',   linewidth=2, label='v1')
ax.quiver(0,0,0, *v2, color='blue',  linewidth=2, label='v2')
ax.quiver(0,0,0, *v3, color='magenta', linewidth=2, label='v3')

# Base parallelogram
ax.add_collection3d(Poly3DCollection([base_face], facecolor='cyan', alpha=0.25))
ax.add_collection3d(Line3DCollection(edges, colors='black', linestyles='--', linewidths=0.8))

# Normal arrow from face center (color by sign)
normal_color = 'tab:green' if signed_volume >= 0 else 'tab:red'
ax.quiver(face_center[0], face_center[1], face_center[2],
          normal_vec[0], normal_vec[1], normal_vec[2],
          color=normal_color, linewidth=3, label='normal (v1  v2)')

# Orientation arc (v1 -> v2)
ax.plot(arc_pts[:,0], arc_pts[:,1], arc_pts[:,2], color='k', linewidth=1.5)
# small arrow tip for the arc (draw last short segment thicker)
if len(arc_pts) >= 2:
    ax.plot(arc_pts[-2:,0], arc_pts[-2:,1], arc_pts[-2:,2], color='k', linewidth=3)


ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('z')
ax.set_title(
    f"Scalar Triple Product (signed volume) = {signed_volume:.3f}\n"
    f"Geometric Volume = {volume:.3f}   |   Orientation: {orientation}"
)
ax.legend(loc='upper left')


all_pts = np.vstack([np.zeros(3), v1, v2, v3, P12, face_center + normal_vec])
mins = all_pts.min(axis=0) - 0.5
maxs = all_pts.max(axis=0) + 0.5
ax.set_xlim(mins[0], maxs[0]); ax.set_ylim(mins[1], maxs[1]); ax.set_zlim(mins[2], maxs[2])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


v1 = np.array([1, 2, 3])
v2 = np.array([2, -1, 1])
v3 = np.array([0, -1, 4])


dot13 = float(np.dot(v1, v3))
dot12 = float(np.dot(v1, v2))
term_A = dot13 * v2                   # (v1v3) v2
term_B = dot12 * v3                   # (v1v2) v3
v_triple = np.cross(v1, np.cross(v2, v3))
rhs = term_A - term_B                 # should equal v_triple

print("v1  (v2  v3) =", v_triple)
print("(v1v3)v2 - (v1v2)v3 =", rhs)
print("Difference =", v_triple - rhs)


def auto_limits(vectors, pad=0.3):
    arr = np.vstack(vectors)
    mins = arr.min(axis=0)
    maxs = arr.max(axis=0)
    ranges = maxs - mins
    max_range = max(ranges.max(), 1.0)
    center = (maxs + mins) / 2
    lo = center - (1+pad) * max_range/2
    hi = center + (1+pad) * max_range/2
    return lo, hi

all_vecs = [v1, v2, v3, term_A, term_B, v_triple]
(lo, hi) = auto_limits(all_vecs)

# --- Plot ---
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

def draw(vec, label, color, lw=2, alpha=1.0):
    ax.quiver(0,0,0, *vec, color=color, linewidth=lw, alpha=alpha,
              arrow_length_ratio=0.1, normalize=False, label=label)

# Original vectors
draw(v1, "v1", "red")
draw(v2, "v2", "blue")
draw(v3, "v3", "green")

def draw1(vec, label, color, lw=1, alpha=0.5):
    ax.quiver(0,0,0, *vec, color=color, linewidth=lw, alpha=alpha,
              arrow_length_ratio=0.1, normalize=False, label=label)

# Identity terms
draw1(term_A, "(v1v3)v2",  "orange")
draw1(term_B, "(v1v2)v3", "cyan")

# Vector triple product
draw(v_triple, "v1  (v2  v3)", "purple", lw=3)

# Axes
ax.set_xlim(lo[0], hi[0]); ax.set_ylim(lo[1], hi[1]); ax.set_zlim(lo[2], hi[2])
ax.set_xlabel("x-axis"); ax.set_ylabel("y-axis"); ax.set_zlabel("z-axis")
ax.set_title("Vector Triple Product and Identity Components")
ax.legend(loc="upper left")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


v1 = np.array([1, 2])
v2 = np.array([3, -1])


a1, a2 = 2, -1

# Linear combination
result = a1*v1 + a2*v2

text = str(a1) + " * " + str(v1) + " + " + str(a2) + " * " + str(v2) + " = " + str(result)



plt.figure(figsize=(6,6))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Original vectors
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='green', label='v2')

# Scaled vectors
scaled_v1 = a1 * v1
scaled_v2 = a2 * v2

plt.quiver(0, 0, scaled_v1[0], scaled_v1[1], angles='xy', scale_units='xy', scale=1, color='blue', alpha=0.5)
plt.quiver(0, 0, scaled_v2[0], scaled_v2[1], angles='xy', scale_units='xy', scale=1, color='green', alpha=0.5)

# Dashed lines
plt.plot([0, scaled_v1[0]], [0, scaled_v1[1]], 'b--', linewidth=1, label='2*v1')
plt.plot([0, scaled_v2[0]], [0, scaled_v2[1]], 'g--', linewidth=1, label='-1*v2')

# Result vector
plt.quiver(0, 0, result[0], result[1], angles='xy', scale_units='xy', scale=1, color='red', label=text)

# Parallelogram construction
plt.plot([scaled_v1[0], result[0]], [scaled_v1[1], result[1]], 'k--')
plt.plot([scaled_v2[0], result[0]], [scaled_v2[1], result[1]], 'k--')

# Mark points
plt.scatter(*scaled_v1, color='blue')
plt.scatter(*scaled_v2, color='green')
plt.scatter(*result, color='red')

plt.xlim(-5, 5)
plt.ylim(-5, 6)
plt.grid(True, linestyle='--', alpha=0.3)
plt.legend()
plt.title("Linear Combination")
plt.show()
```



* Vectors are **linearly independent** if **no vector** can be written as a linear combination of the others.
* In $\mathbb{R}^2$, two non-collinear vectors are independent.
* In $\mathbb{R}^3$, three non-coplanar vectors are independent.

 Linearly Independent Vectors in $\mathbb{R}^3$

$$
v_1 = \begin{bmatrix}1 \\ 0 \\ 0\end{bmatrix}, \quad
v_2 = \begin{bmatrix}0 \\ 1 \\ 0\end{bmatrix}, \quad
v_3 = \begin{bmatrix}0 \\ 0 \\ 1\end{bmatrix}
$$

These are the **standard basis vectors** in 3D - clearly independent.




```python
import matplotlib.pyplot as plt
import numpy as np

v1 = np.array([1, 0, 0])
v2 = np.array([0, 1, 0])
v3 = np.array([0, 0, 1])


fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection='3d')


ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='red', label='v1 = (1,0,0)')
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='green', label='v2 = (0,1,0)')
ax.quiver(0, 0, 0, v3[0], v3[1], v3[2], color='blue', label='v3 = (0,0,1)')


ax.set_xlim([0, 1.5])
ax.set_ylim([0, 1.5])
ax.set_zlim([0, 1.5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
ax.set_title("Linearly Independent Vectors in 3D")

plt.show()
```

**Many possible sets** of linearly independent vectors in $\mathbb{R}^3$.

Another example that is clearly not aligned with the standard basis:

$$
v_1 = \begin{bmatrix}1 \\ 1 \\ 0\end{bmatrix}, \quad
v_2 = \begin{bmatrix}0 \\ 1 \\ 1\end{bmatrix}, \quad
v_3 = \begin{bmatrix}1 \\ 0 \\ 1\end{bmatrix}
$$



* If one try to solve $a v_1 + b v_2 + c v_3 = 0$, the only solution would be $a=b=c=0$.

That proves they are linearly independent in $\mathbb{R}^3$.




$$
v_1 = (1,1,0), \quad v_2 = (0,1,1), \quad v_3 = (1,0,1)
$$

These are linearly independent, but they are not orthogonal (their dot products are not zero).

**Orthogonal Vectors**


$$
v_1 = (1,0,0), \quad v_2 = (0,1,0), \quad v_3 = (0,0,1)
$$

They satisfy:

* $v_1 \cdot v_2 = 0$
* $v_1 \cdot v_3 = 0$
* $v_2 \cdot v_3 = 0$

They are mutually  orthogonal
(and orthonormal if length = 1).




---
 Case of 2 vectors in $\mathbb{R}^n$:

 If $v_1$ and $v_2$ are **linearly dependent**, then

  $$
  v_2 = c \, v_1
  $$

  for some scalar $c$.
* Geometrically  they are **parallel (or anti-parallel)**.
  In **2-vector case**, dependence means parallel.



2. Case of 3 or more vectors:

They need not be parallel. They can lie in the same **plane or lower-dimensional subspace**.

Example in $\mathbb{R}^3$:

$$
v_1 = (1,0,0), \quad v_2 = (0,1,0), \quad v_3 = (1,1,0)
$$

Here,

$$
v_3 = v_1 + v_2
$$

So $v_1,v_2,v_3$ are **linearly dependent**, but they are re not all parallel  they just lie in the same plane (the XY-plane).



```python
import matplotlib.pyplot as plt
import numpy as np

v1 = np.array([2, 1])
v2 = np.array([6, 3])


plt.figure(figsize=(6,6))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)



plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='red', label='v1 = (2,1)')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v2 = (6,3) = 3*v1')

plt.xlim(-1, 7)
plt.ylim(-1, 5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.title("Linearly Dependent Vectors in R (Parallel)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import matplotlib.lines as mlines

# Vectors (same as previous example)
v1 = np.array([2.0, 0.0, 1.0])
v2 = np.array([0.0, 2.0, 1.0])
v3 = np.array([1.0, 1.0, 3.0])
w  = 0.6 * v1 + 0.3 * v2         # (on plane span{v1, v2})


fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

u = np.linspace(0, 1, 15)
v = np.linspace(0, 1, 15)
U, V = np.meshgrid(u, v)
X = U * v1[0] + V * v2[0]
Y = U * v1[1] + V * v2[1]
Z = U * v1[2] + V * v2[2]
ax.plot_surface(X, Y, Z, alpha=0.25, linewidth=0, antialiased=True, color='lightblue')


def draw(ax, vec, color, label):
    ax.quiver(0, 0, 0, vec[0], vec[1], vec[2], arrow_length_ratio=0.08, color=color)
    ax.scatter([vec[0]], [vec[1]], [vec[2]], s=35, color=color)
    ax.text(vec[0], vec[1], vec[2], f"  {label}", fontsize=11, color=color)


draw(ax, v1, 'red', 'v1')
draw(ax, v2, 'magenta', 'v2')
draw(ax, v3, 'green', 'v3 (off-plane)')
draw(ax,  w, 'black', 'w = 0.6 v1 + 0.3 v2 (on plane)')

ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], linestyle='--', linewidth=1, color='red', alpha=0.6)
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], linestyle='--', linewidth=1, color='magenta', alpha=0.6)
ax.plot([v1[0], v1[0]+v2[0]], [v1[1], v1[1]+v2[1]], [v1[2], v1[2]+v2[2]], linestyle='--', linewidth=1, color='gray', alpha=0.6)
ax.plot([v2[0], v1[0]+v2[0]], [v2[1], v1[1]+v2[1]], [v2[2], v1[2]+v2[2]], linestyle='--', linewidth=1, color='gray', alpha=0.6)


proxy_v1 = mlines.Line2D([], [], color='red', marker='>', linestyle='None', markersize=8, label='v1 ')
proxy_v2 = mlines.Line2D([], [], color='magenta', marker='>', linestyle='None', markersize=8, label='v2 ')
proxy_v3 = mlines.Line2D([], [], color='green', marker='>', linestyle='None', markersize=8, label='v3 ')
proxy_w  = mlines.Line2D([], [], color='black', marker='>', linestyle='None', markersize=8, label='w ')
proxy_plane = mlines.Line2D([], [], color='lightblue', linestyle='-', linewidth=8, alpha=0.25, label='Plane span{v1, v2}')
ax.legend(handles=[proxy_v1, proxy_v2, proxy_v3, proxy_w, proxy_plane], loc='upper left')


m = max(np.linalg.norm(v1), np.linalg.norm(v2), np.linalg.norm(v3)) * 1.4
ax.set_xlim(0, m); ax.set_ylim(0, m); ax.set_zlim(0, m)
ax.set_xlabel('X'); ax.set_ylabel('Y'); ax.set_zlabel('Z')
ax.set_title("w lies in the plane span{v1, v2}; v1=red, v2=magenta, v3=green, w=black")

plt.show()
```
