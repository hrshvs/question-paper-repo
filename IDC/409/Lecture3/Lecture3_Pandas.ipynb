Pivoting data frame
--
The **pivot()** function is used to reshape a given DataFrame object organized by given index/column values

**DataFrame.pivot(self, index=None, columns=None, values=None)**

where

*index*, specifies column that will be used to make new frame's index. If no column is specified, existing index is used.

*columns* specify columns that will be used to make new frame's columns.

*values* specify column(s)to be used for populating new frame's values. if nothing is specified, all remaining columns will be used.

**pivot()** returns a reshaped DataFrame.

ValueError will be raised of there are any duplicate values in the data frame

Pivot is basically a reshape operation - it takes ones long/tidy data and turn into a wide table.

It does three things:

> Pick a column to become new row index (unique lables for the left-hand side of your table.)

> Pick a column whose unique values new columns (spreads these values across the top as headers.)

>  Pick a column whose values fill the cells (the data that will go into the table body.)

It does not aggregate. It just rearranges the data.

This means there must be exactly one row for each (index, column) pair â€” no duplicates.

If duplicates exist, you get a ValueError and must use pivot_table instead.

````python
import pandas as pd

data = pd.DataFrame({
    "Name": [
        "Captain Vyom", "Captain Vyom",
        "Captain Blaze", "Captain Blaze",
        "Captain Cyborg", "Captain Cyborg",
        "Captain Maya", "Captain Maya",
        "Captain Chaya", "Captain Chaya"
    ],
    "Attribute": [
        "Age", "Power",
        "Age", "Power",
        "Age", "Power",
        "Age", "Power",
        "Age", "Power"
    ],
    "Value": [
        27, "Telepathy",
        35, "Fire Control",
        40, "Cybernetic Strength",
        30, "Illusion Casting",
        28, "Shadow Manipulation"
    ],
    "Role": [
        "Leader", "Leader",
        "Pilot", "Pilot",
        "Tactician", "Tactician",
        "Strategist", "Strategist",
        "Infiltrator", "Infiltrator"
    ],
    "Actor": [
        "Milind Soman", "Milind Soman",
        "Vinod", "Vinod",
        "Rajat", "Rajat",
        "Priya", "Priya",
        "Anita", "Anita"
    ]
})

data
````

````python
pivoted = data.pivot(index=["Name", "Role", "Actor"], columns="Attribute", values="Value")
#pivoted = pivoted.reset_index()   optional: bring index back as columns
pivoted
````

We can a twist with pivot_table so one can see how it works if two Captains share the same name or role and we need to aggregate data

 We add twist using Captain Vyom + Maya + Chaya (and friends). we'll introduce duplicate Ages (conflicting sources), which breaks pivot, then fix it with pivot_table.

````python
data = pd.DataFrame({
    "Name": [
        "Captain Vyom","Captain Vyom","Captain Vyom",   # <-- Age duplicated (27 & 28)
        "Captain Blaze","Captain Blaze",
        "Captain Cyborg","Captain Cyborg",
        "Captain Maya","Captain Maya","Captain Maya",   # <-- Age duplicated (30 & 31)
        "Captain Chaya","Captain Chaya"
    ],
    "Attribute": [
        "Age","Age","Power",
        "Age","Power",
        "Age","Power",
        "Age","Age","Power",
        "Age","Power"
    ],
    "Value": [
        27, 28, "Telepathy",
        35, "Fire Control",
        40, "Cybernetic Strength",
        30, 31, "Illusion Casting",
        28, "Shadow Manipulation"
    ],
    "Role": [
        "Leader","Leader","Leader",
        "Pilot","Pilot",
        "Tactician","Tactician",
        "Strategist","Strategist","Strategist",
        "Infiltrator","Infiltrator"
    ],
    "Actor": [
        "Milind Soman","Milind Soman","Milind Soman",
        "Vinod","Vinod",
        "Rajat","Rajat",
        "Priya","Priya","Priya",
        "Anita","Anita"
    ]
})

data

````

````python
# 1) Show why plain pivot breaks
data.pivot(index=["Name","Role","Actor"], columns="Attribute", values="Value")
#  ValueError: Index contains duplicate entries (because Age repeats for Vyom & Maya)
````

For each Name (row) and Attribute (column), it expects exactly one value from Value

If there's more than one, Pandas doesn't know whether to take the first, the last, the mean, or something else â€” so instead of guessing, it raises:




````python
# 2) Use pivot_table to AGGREGATE duplicates
# Strategy:
#   - For Age (numeric duplicates): take the mean
#   - For Power (strings, single per character): just take the first

pt_age = pd.pivot_table(
    data.query("Attribute == 'Age'"),
    index=["Name","Role","Actor"],
    columns="Attribute",
    values="Value",
    aggfunc="mean"         # averages duplicate Ages
)

pt_power = pd.pivot_table(
    data.query("Attribute == 'Power'"),
    index=["Name","Role","Actor"],
    columns="Attribute",
    values="Value",
    aggfunc="first"        # grabs the single Power value
)

final = pt_age.join(pt_power, how="outer").reset_index()
final
````

Sorting
--
One can sort data either in ascending or descending order.

Both numberica and text data can be sorted in a particular oder.
* Pandas  **sort_values()** function sorts a data frame in ascending or descending order of passed column

**DataFrame.sort_Values(by, axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last')**


* **by str or list of str**
Name or list of names to sort by.
  * if axis is 0 or â€˜indexâ€™ then by may contain index levels and/or column labels.
  * if axis is 1 or â€˜columnsâ€™ then by may contain column levels and/or index labels.

* **axis : {0 or â€˜indexâ€™, 1 or â€˜columnsâ€™}, default 0**
  * Axis to be sorted.

* **ascending :bool or list of bool, default True**
  *  Sort ascending vs. descending. Specify list for multiple sort orders. If this is a list of bools, must match the length of the by.

* **inplace : bool, default False**
  * If True, perform operation in-place.

* **kind :{â€˜quicksortâ€™, â€˜mergesortâ€™, â€˜heapsortâ€™, â€˜stableâ€™}, default â€˜quicksortâ€™**
  * Choice of sorting algorithm. See also numpy.sort() for more information. mergesort and stable are the only stable algorithms. For DataFrames, this option is only applied when sorting on a single column or label.
    
    
* **na_position : {â€˜firstâ€™, â€˜lastâ€™}, default â€˜lastâ€™**
  * Puts NaNs at the beginning if first; last puts NaNs at the end

````python
Cars= {'Brand': ['BMW M50', 'Tesla Model 3', 'BMW M3 Series', 'Mercedes G500', 'Mercedes E200', 'Jaguar'], 'Price':[ 50000, 80000, 70000, 45000, 80000, 55000], 'Year' : [2019, 2022, 2018, 2020, 2023, 2021]}
df_car= pd.DataFrame(Cars, columns=['Brand', 'Price','Year'] )
print(df_car)

print(' \n\n Now, I will sort by price \n')
df_car.sort_values(by='Price', inplace=True, ascending=False)
print(df_car)

print(' \n\n Now, I will sort by Name \n')
df_car.sort_values(by='Brand', inplace=True, ascending=False)
print(df_car)

print(' \n\n Now, I will sort by price let me make inplace=FALSE \n')
df_car_2 =df_car.sort_values(by='Price', inplace=False, ascending=False)
print('\n the original has not changed \n')
print(df_car)
print('\n sorted \n')
print(df_car_2)
````

**fill_value**
When working with series and DataDrame, one can use  arithmetic functions to input a **fill_value** that can be used to substitute missing values.

If one data frame has a NaN at a particular location and a non-NAN value at the corresponding location in other data drame, then it is better to replace NaN.

**Fill existing missing (NaN) values, and any new element needed for successful DataFrame alignment, with this value before computation. If data in both corresponding DataFrame locations is missing the result will be missing.**



````python
df1 = pd.DataFrame({'one':pd.Series([1,2,3,4,5], index=['a','b','c','d','e']), 'two':pd.Series([6,7,8,9,10],index=['a','b','c','d','e']), 'three':pd.Series( [11,12,13,14,15],index=['a','b','c','d','e']) })
print(' Data frame 1 \n', df1)
df2 = pd.DataFrame({'one':pd.Series([1,2,3], index=['a','b','c']), 'two':pd.Series([4,5,6,7],index=['a','b','c','d']), 'three':pd.Series( [8,9,10,12],index=['a','b','c','e']) })
print(' Data frame 2 \n', df2)

print(" \n\nAdding Data Frame with missing values")
print(df1.add(df2))

print(" \n\n Adding Data Frame with missing values")

print(df1.add(df2, fill_value=987656789))
````

**Combining data frames**
--
One can combine two data frames.

* **merge()** combines datasets on common column or index or both
* **concat()** combines datasets accross rows or columns
* **join()** combines datasets on key column or index

**merge()**  is the most flexible of the three operations.

 merge() is most useful when you want to combine rows that share data.

 One can achieve both many-to-one and many-to-many joins with merge().
 * In a many-to-one join, one of your datasets will have many rows in the merge column that repeat the same values. The merge column in the other dataset won't have repeated values.
 * In a many-to-many join, both of your merge columns will have repeated values.

**pd.merge(left, right, how="inner", on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=True)**


One need two reguired arguments
* left DataFrame
* right DataFrame

**how**, specifies the kind of merge to make. Default value is *inner*  but other possible options are "outer", "left", and "right"

**on**, indicates columns or indices on whihc join operation will be performed
* By default  if left_index and right_index are false, then columns from two dataFrame that share names will be used as join keys.
* If one uses on, then the specified column or index must be present in both objects

**combine_first()**

Combine two DataFrame objects by filling null values in one DataFrame with non-null values from other DataFrame.

The row and column indexes of the resulting DataFrame will be the union of the two.

The resulting dataframe contains the â€˜firstâ€™ dataframe values and overrides the second one values where both *first.loc[index, col]* and *second.loc[index, col]* are not missing values, upon calling *first.combine_first(second)*.

````python
import pandas as pd
df1 = pd.DataFrame({'A': [None, 0], 'B': [None, 4]})
df2 = pd.DataFrame({'A': [1, 2], 'B': [3, 5]})
print(df1)
print(df2)
df1.combine_first(df2)
````

````python
df2 = pd.DataFrame({'A': [None, 0], 'B': [None, 4]})
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 5]})
print(df1)
print(df2)
df1.combine_first(df2)
````

````python
df1 = pd.DataFrame({'A': [None, 0], 'B': [4, None]})

df2 = pd.DataFrame({'B': [3, 5], 'C': [1, 2]}, index=[1, 2])
print(df1)
print(df2)
df1.combine_first(df2)
````

Function application
--

While coding, one might have to apply functions to Pandas objects.

Pandas provide
* **pipe()** method applies function on Table
* **apply()** method applies function row or column
* **applymap()** applies dunction to each elements



````python
import pandas as pd
def madd(num1, num2):
  return num1+num2

def mmul(num1,num2):
  return num1*num2

d = {
   'Cost':pd.Series([25,26,25,23,30,29,23,34,40,30,51,46]),
   'Rating':pd.Series([4.23,3.24,3.98,2.56,3.20,4.6,3.8,3.78,2.98,4.80,4.10,3.65])
}


#Create a DataFrame
df = pd.DataFrame(d)
print(df)
print('df.pipe(madd,2200)')
df.pipe(madd,2200)
````

If one has to apply mutiple functions

````python
print('df.pipe(madd,3000).pipe(mmul,10)')
df.pipe(madd,3000).pipe(mmul,10)
````
